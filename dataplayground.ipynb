```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create sample real estate data, will replace with actual data later
np.random.seed(42)  # For reproducibility

# Generate data for houses with 1-5 bedrooms
bedrooms = np.random.randint(1, 6, 50)  # 50 houses

# Generate prices based on bedrooms with some randomness
# Base price: $200,000, add $50,000 per bedroom, plus random variation
base_price = 200000
price_per_bedroom = 50000
random_variation = np.random.normal(0, 30000, 50)  # Random variation

prices = base_price + (bedrooms * price_per_bedroom) + random_variation

# Create DataFrame
df = pd.DataFrame({
    'bedrooms': bedrooms,
    'price': prices
})

print("Sample of the data:")
print(df.head(10))
print(f"\nTotal houses: {len(df)}")
```

```python
# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['bedrooms'], df['price'], alpha=0.6, s=100, c='steelblue', edgecolors='black')

# Add labels and title
plt.xlabel('Number of Bedrooms', fontsize=12)
plt.ylabel('Selling Price ($)', fontsize=12)
plt.title('House Selling Price vs Number of Bedrooms', fontsize=14, fontweight='bold')

# Format y-axis to show prices in a readable format
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

# Add grid for better readability
plt.grid(True, alpha=0.3)

# Set x-axis to show only integer bedroom counts
plt.xticks(range(1, 6))

plt.tight_layout()
plt.show()
```

```python
# Calculate average price per bedroom count
avg_prices = df.groupby('bedrooms')['price'].mean().sort_index()

print("Average Selling Price by Number of Bedrooms:")
print("-" * 50)
for bedrooms, avg_price in avg_prices.items():
    print(f"{bedrooms} bedroom(s): ${avg_price:,.2f}")
```
